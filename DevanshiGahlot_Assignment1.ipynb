# Problem 1 - Which industry has the maximum brands whose Brand_Revenue > 80 Million (Brand revenue is in millions)


```python
#I have attempted this question in two parts:
#First I have filtered the Industries whose "Brand_Revenue"is Greater than 80.
new=df.loc[df["Brand_Revenue"]>80]
new
```

```python
#Second I have grouped the industries and evaluated the count and on the basis of the count (found the industry corresponding to the maximum count) 
print(new.groupby(by="Industry")["Brand"].count())
print(new.groupby(by="Industry")["Brand"].count().max())
print(new["Industry"][new.groupby(by="Industry")["Brand"].count().max()])
```

# Problem 2 - List the companies whose Company_Advertising budget is < 3 but Brand_Value is > 20

```python
df.loc[(df['Company_Advertising']<3) & (df['Brand_Value'] > 20)][['Brand','Company_Advertising',"Brand_Value"]]
```

# Problem 3 - Find out the top 5 Technology Brand that showed maximum positive jump in their position (Use Change_LastYr)


```python
top=df.loc[(df["Industry"]=="Technology")]
top_1=top.nlargest(5,columns=["Change_LastYr"])
top_1.loc[:,["Brand","Change_LastYr","Industry"]]
```
